```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

### Exploratory Data Analysis

```python
# read dataset
df = pd.read_csv('concrete_data.csv')
df.head()
```

```python
df.describe()
```

```python
# plot the correlation plot for the df
df.corr().style.background_gradient(cmap='coolwarm')
```

### Splitting the data into independent and dependent variables

```python
X =  df.iloc[:,:-1]
y = df.iloc[:,-1]
```

## Bulid a Keras Sequential  baseline model

```python
def init_model(n = 10):
  model  = keras.Sequential([
    keras.layers.Dense(n, activation=tf.nn.relu, input_shape=(X.shape[1],)),
    keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam', loss='mse')
  return model
```

```python
def train_eval(model, X, y, n_epochs = 50):
  # Splitting the dataset into the Training set and Test set
  X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
  history = model.fit(X_train, y_train, epochs=n_epochs, verbose=0)
  score = model.evaluate(X_test, y_test, verbose=0)
  return score
```

```python
history = {'iteration' : [], 'test_mse': []}

for i in range(50):
  model = init_model()
  score = train_eval(model, X,y, n_epochs=50)
  history['iteration'].append(i)
  history['test_mse'].append(score)
  # print(f"Iteration: {i}, and test MSE: {score}")
```

## Plot the History

```python
tdf = pd.DataFrame(history)
tdf.head()
```

```python
tdf.boxplot(column=['test_mse'], grid=False)
plt.title("Baseline Model")
plt.show()
```

```python
# Reporting the mean and standard deviations
print("Mean=", tdf['test_mse'].mean(), "and Standard deviation=", tdf['test_mse'].std())
```

## B. Using normalization

## D. Increasing the number of hidden layers

```python
def init_deeper_model():
  model  = keras.Sequential([
    keras.layers.Dense(10, activation=tf.nn.relu, input_shape=(X.shape[1],)),
    keras.layers.Dense(10, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.relu),
    keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam', loss='mse')
  return model
```

```python
history = {'iteration' : [], 'test_mse': []}

for i in range(50):
  model = init_deeper_model()
  score = train_eval(model, X,y, n_epochs=50)
  history['iteration'].append(i)
  history['test_mse'].append(score)
  # print(f"Iteration: {i}, and test MSE: {score}")
```

```python
tdf = pd.DataFrame(history)
tdf.head()
```

```python
tdf.boxplot(column=['test_mse'], grid=False)
plt.title("Deep model with 3 hidden layers")
plt.show()
```

```python
# Reporting the mean and standard deviations
print("Mean=", tdf['test_mse'].mean(), "and Standard deviation=", tdf['test_mse'].std())
```
