```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

### Exploratory Data Analysis

```python
# read dataset
df = pd.read_csv('concrete_data.csv')
df.head()
```

```python
df.describe()
```

```python
# plot the correlation plot for the df
df.corr().style.background_gradient(cmap='coolwarm')
```

### Standize the data and split it into training and testing sets

```python
X =  df.iloc[:,:-1]
y = df.iloc[:,-1]

# Standardize the dataset
scaler = StandardScaler()
X = scaler.fit_transform(X)
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

## Bulid a Keras Sequential model

```python
model  = keras.Sequential([
    keras.layers.Dense(100, activation=tf.nn.relu, input_shape=(X_train.shape[1],)),
    keras.layers.Dense(100, activation=tf.nn.relu),
    keras.layers.Dense(100, activation=tf.nn.relu),
    keras.layers.Dense(100, activation=tf.nn.relu),
    keras.layers.Dense(100, activation=tf.nn.relu),
    # keras.layers.Dense(100, activation=tf.nn.relu),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```

```python
# Train and evaluate the model
n_epochs = 100
history = model.fit(X_train, y_train, validation_split=0.2, epochs=n_epochs, verbose=0)
```

## Plot the History

```python
epochs = np.arange(0,n_epochs)
plt.style.use("ggplot")
plt.figure(1, figsize=(10,6))
plt.plot(epochs, history.history["loss"], label="train_loss", marker='o')
plt.plot(epochs, history.history["val_loss"], label="val_loss", marker='d')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## Perform Test Evaluation

```python
# evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Scores {scores}")
```

```python
pred = model.predict(X_test)
```

```python
x = np.arange(0, y_test.shape[0])
plt.figure(2, figsize=(10,6))
plt.plot(x[:50:], y_test[:50:], marker='o', label='Ground Truth')
plt.plot(x[:50], pred[:50:], marker='d', label='Prediction')
plt.ylabel("Predicted")
plt.xlabel("Sample")
plt.legend()
plt.show()
```
